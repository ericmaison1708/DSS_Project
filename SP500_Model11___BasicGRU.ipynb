```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date
```

```python
start = '2010-01-01'
end = date.today().strftime("%Y-%m-%d")

df = yf.download('AAPL', start=start, end=end)
df.head()
```

```python
df.tail()
```

```python
df = df.reset_index()
df.head()
```

```python
df = df.drop(['Date', 'Adj Close'], axis=1)
df.head()
```

```python
plt.plot(df.Close)
```

```python
ma100 = df.Close.rolling(100).mean()
ma100
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.Close)
plt.plot(ma100, 'r')
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.Close)
plt.plot(ma100, 'r')
plt.plot(ma200, 'g')
```

```python
# Chia tp d liu thnh training_set v testing_set
n1 = int(len(df)*0.80)
n2 = int(len(df)*0.10)

data_training = pd.DataFrame(df['Close'][0:n1])
data_validating = pd.DataFrame(df['Close'][n1:(n1+n2)])
data_testing = pd.DataFrame(df['Close'][(n1+n2):])

print(data_training.shape)
print(data_validating.shape)
print(data_testing.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)    
```

```python
x_train.shape
```

```python
days = data_training.tail(100)
data_validating = pd.concat([days, data_validating], ignore_index=True)
```

```python
data_valiadating_array = scaler.fit_transform(data_validating)
data_valiadating_array
```

```python
x_val = []
y_val = []

for i in range(100, data_valiadating_array.shape[0]):
    x_val.append(data_valiadating_array[i-100:i])
    y_val.append(data_valiadating_array[i, 0])

x_val, y_val = np.array(x_val), np.array(y_val)    
```

```python
x_val.shape
```

```python
# ML Model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
import warnings

tf.get_logger().setLevel('ERROR')

warnings.filterwarnings('ignore')

# Define your model
model = Sequential([
    layers.Input(shape=(100, 1)),
    LSTM(64),
    Dense(1)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
model.fit(x_train, y_train, validation_data=(x_val, y_val),  epochs=100)
```

```python
model.save(r'D:\20232\MI4216 H h tr quyt nh\Project_DSS\Project\Models\AAPL_Model1___BasicLSTM.keras')
```

```python
data_testing.head()
```

```python
past_100_days = data_validating.tail(100)
```

```python
final_df = pd.concat([past_100_days, data_testing], ignore_index=True)
```

```python
input_data = scaler.fit_transform(final_df)
input_data
```

```python
input_data.shape
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)    
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# Making predictions
y_predicted = model.predict(x_test)
```

```python
y_predicted = scaler.inverse_transform(y_predicted.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)
rmse = np.sqrt(mse)
mape = (np.abs((y_test - y_predicted) / y_test).mean()) * 100
r2 = r2_score(y_test, y_predicted)

# nh gi m hnh
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}%')
print(f'R: {r2}')
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test,'b', label='Original Price')
plt.plot(y_predicted, 'r', label='Predicted Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
