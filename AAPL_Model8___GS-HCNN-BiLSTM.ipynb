```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import date
```

```python
start = '2010-01-01'
end = date.today().strftime("%Y-%m-%d")

df = yf.download('AAPL', start=start, end=end)
df.head()
```

```python
# Chia tập dữ liệu thành training_set và testing_set
n1 = int(len(df)*0.70)
n2 = int(len(df)*0.15)

data_training = pd.DataFrame(df['Close'][0:n1])
data_validating = pd.DataFrame(df['Close'][n1:(n1+n2)])
data_testing = pd.DataFrame(df['Close'][(n1+n2):])

print(data_training.shape)
print(data_validating.shape)
print(data_testing.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)    
```

```python
x_train.shape
```

```python
days = data_training.tail(100)
data_validating = pd.concat([days, data_validating], ignore_index=True)
```

```python
data_valiadating_array = scaler.fit_transform(data_validating)
data_valiadating_array
```

```python
x_val = []
y_val = []

for i in range(100, data_valiadating_array.shape[0]):
    x_val.append(data_valiadating_array[i-100:i])
    y_val.append(data_valiadating_array[i, 0])

x_val, y_val = np.array(x_val), np.array(y_val)    
```

```python
x_val.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dropout, Conv1D, MaxPooling1D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from keras_tuner import HyperParameters, GridSearch

# Define the model architecture
def build_model(hp):
    model = Sequential()
    
    # Bidirectional LSTM layer
    model.add(Bidirectional(LSTM(hp.Int('units', min_value=32, max_value=512, step=32), return_sequences=True), input_shape=(x_train.shape[1], 1)))
    
    # Dropout layer
    model.add(Dropout(hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)))
    
    # Conv1D layer
    model.add(Conv1D(hp.Int('filters', min_value=32, max_value=512, step=32), kernel_size=3, activation='relu'))
    
    # MaxPooling1D layer
    model.add(MaxPooling1D(pool_size=2))
    
    # Flatten layer
    model.add(Flatten())
    
    # Dense layers
    model.add(Dense(hp.Int('dense_units', min_value=32, max_value=512, step=32), activation='relu'))
    model.add(Dense(hp.Int('dense_units', min_value=32, max_value=512, step=32), activation='relu'))
    
    # Output layer
    model.add(Dense(1, activation='relu'))
    
    # Compile the model
    optimizer = Adam(learning_rate=hp.Choice('learning_rate', values=[0.01, 0.001, 0.0001]))
    model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mean_absolute_error'])
    
    return model

# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_mean_absolute_error', patience=5, restore_best_weights=True, mode='min')

# Create HyperParameters object
hp = HyperParameters()

# Create GridSearch tuner
tuner = GridSearch(
    build_model,
    hyperparameters=hp,
    objective='val_mean_absolute_error',
    max_trials=5,
    directory='grid_search',
    project_name='AAPL_Model8___GS-HCNN-BiLSTM')

# Perform hyperparameter tuning
tuner.search(x=x_train,
             y=y_train,
             epochs=50,
             batch_size=32,
             validation_data=(x_val, y_val),
             callbacks=[early_stopping])

# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)    
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# Making predictions
y_predicted = best_model.predict(x_test)
```

```python
y_predicted = scaler.inverse_transform(y_predicted.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_predicted)
mse = mean_squared_error(y_test, y_predicted)
rmse = np.sqrt(mse)
mape = (np.abs((y_test - y_predicted) / y_test).mean()) * 100
r2 = r2_score(y_test, y_predicted)

# Đánh giá mô hình
print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}%')
print(f'R²: {r2}')
```
