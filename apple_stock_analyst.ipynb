```python
# Import relevant library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Đọc dữ liệu từ tệp CSV
apple_data = pd.read_csv('../Data/AAPL.csv')
market_data = pd.read_csv('../Data/GSCP.csv')
```

```python
# Chuyển đổi cột 'Date' thành định dạng datetime
apple_data['Date'] = pd.to_datetime(apple_data['Date'])
apple_data
```

```python
# Tính toán thống kê mô tả cho các cột số
statistics = apple_data.describe()
statistics
```

```python
# Thiết lập các biểu đồ
plt.style.use('seaborn-v0_8-darkgrid')

# Vẽ histogram cho các cột giá trị
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
fig.suptitle('Distribution of Stock Prices and Volume')

# Histogram cho các cột giá trị
sns.histplot(apple_data['Open'], bins=50, ax=axs[0, 0], kde=True).set(title='Open Price')
sns.histplot(apple_data['High'], bins=50, ax=axs[0, 1], kde=True).set(title='High Price')
sns.histplot(apple_data['Low'], bins=50, ax=axs[1, 0], kde=True).set(title='Low Price')
sns.histplot(apple_data['Close'], bins=50, ax=axs[1, 1], kde=True).set(title='Close Price')
sns.histplot(apple_data['Adj Close'], bins=50, ax=axs[2, 0], kde=True).set(title='Adjusted Close Price')
sns.histplot(apple_data['Volume'], bins=50, ax=axs[2, 1], kde=True).set(title='Volume')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Tóm tắt dữ liệu theo năm
apple_data['Year'] = apple_data['Date'].dt.year
yearly_summary = apple_data.groupby('Year').agg({
    'Open': 'mean',
    'High': 'mean',
    'Low': 'mean',
    'Close': 'mean',
    'Adj Close': 'mean',
    'Volume': 'sum'
}).reset_index()
print(yearly_summary)
```

```python
# Tính toán lợi nhuận hàng ngày
apple_data['Daily Return'] = apple_data['Adj Close'].pct_change()

# Histogram lợi nhuận hàng ngày
plt.figure(figsize=(8, 6))
sns.histplot(apple_data['Daily Return'].dropna(), bins=50, kde=True)
plt.title('Daily Return Distribution')
plt.show()
```

```python
def calculate_technical_indicators(df):
    df['SMA'] = df['Close'].rolling(window=20).mean()
    df['RSI'] = 100 - (100 / (1 + df['Close'].diff(1).rolling(window=14).mean() / df['Close'].diff(1).rolling(window=14).mean().abs()))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Bollinger_Upper'] = df['SMA'] + (df['Close'].rolling(window=20).std() * 2)
    df['Bollinger_Lower'] = df['SMA'] - (df['Close'].rolling(window=20).std() * 2)
    df.dropna(inplace=True)
    return df

apple_data = calculate_technical_indicators(apple_data)
market_data = calculate_technical_indicators(market_data)

apple_data
```
